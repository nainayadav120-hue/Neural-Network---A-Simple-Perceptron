```python
#2. Explain the basic architecture and functioning of a Perceptron. What are its limitations?
""" Perceptron is a type of artificial neural network unit that serves as a building block for more complex neural networks. It is a simple, single-layer neural network used for binary classification tasks.

Basic Architecture:
1. Inputs (x): The Perceptron receives one or more input features.
2. Weights (w): Each input is associated with a weight, which determines the importance of the input.
3. Summation: The weighted inputs are summed up.
4. Activation Function: The sum is passed through an activation function (typically a step function) to produce the output.
5. Output (y): The output is a binary value (0 or 1) indicating the predicted class.

Functioning:
1. Initialization: Weights are initialized randomly.
2. Forward Pass: Inputs are multiplied by their corresponding weights and summed up.
3. Activation: The sum is passed through the activation function to produce the output.
4. Error Calculation: The error between the predicted output and the actual output is calculated.
5. Weight Update: The weights are updated based on the error and the learning rate.

Limitations:
1. Linear Separability: The Perceptron can only learn linearly separable patterns. If the data is not linearly separable, the Perceptron will not be able to classify it correctly.
2. Binary Classification: The Perceptron is limited to binary classification tasks.
3. No Hidden Layers: The Perceptron does not have any hidden layers, which limits its ability to learn complex patterns.
4. Sensitive to Noise: The Perceptron is sensitive to noisy data and outliers.

Example:Suppose we want to classify points in a 2D space as either belonging to class A or class B.
If the points are linearly separable (i.e., a straight line can separate the two classes), the Perceptron can learn to classify them correctly.
However, if the points are not linearly separable (e.g., they are intertwined), the Perceptron will not be able to classify them correctly."""

#Code Example (Python):
import numpy as np

class Perceptron:
    def __init__(self, learning_rate=0.01, num_iterations=1000):
        self.learning_rate = learning_rate
        self.num_iterations = num_iterations
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        num_samples, num_features = X.shape
        self.weights = np.zeros(num_features)
        self.bias = 0

        for _ in range(self.num_iterations):
            for idx, x_i in enumerate(X):
                linear_output = np.dot(x_i, self.weights) + self.bias
                y_predicted = np.where(linear_output >= 0, 1, 0)

                update = self.learning_rate * (y[idx] - y_predicted)
                self.weights += update * x_i
                self.bias += update

    def predict(self, X):
        linear_output = np.dot(X, self.weights) + self.bias
        y_predicted = np.where(linear_output >= 0, 1, 0)
        return y_predicted

# Example usage
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 1, 1, 1])  # OR gate
perceptron = Perceptron()
perceptron.fit(X, y)
print(perceptron.predict(X))
```

```python
#3. Describe the purpose of activation function in neural networks. Compare Sigmoid, ReLU, and Tanh functions.
"""Activation Functions in neural networks are used to introduce non-linearity into the model, enabling it to learn complex patterns and relationships in the data. Without activation functions, neural networks would only be able to learn linear relationships.

Purpose of Activation Functions:
1. Introduce Non-Linearity: Allow the model to learn non-linear relationships between inputs and outputs.
2. Enable Deep Learning: Without activation functions, deep neural networks would not be able to learn complex patterns.
3. Improve Model Performance: Activation functions help the model to better fit the data and improve its performance.

Comparison of Sigmoid, ReLU, and Tanh Functions:
1. Sigmoid Function- Formula: σ(x) = 1 / (1 + e^(-x))
- Range: (0, 1)
- Derivative: σ(x)  = σ(x)(1-σ(x))
- Pros:
    - Output can be interpreted as probability.
    - Differentiable.
- Cons:
    - Suffers from vanishing gradient problem (output values are close to 0 or 1, gradients are close to 0).
    - Not zero-centered.

2. ReLU (Rectified Linear Unit) Function- Formula: f(x) = max(0, x)
- Range: [0, ∞)
- Derivative: f'(x) = 1 if x > 0, 0 otherwise
- Pros:
    - Computationally efficient.
    - Does not suffer from vanishing gradient problem.
    - Sparse output (some neurons output 0).
    - Cons:
    - Not differentiable at x = 0.
    - Dying ReLU problem (neurons output 0 for all inputs).

3. Tanh Function- Formula: tanh(x) = 2 / (1 + e^(-2x)) - 1
- Range: (-1, 1)
- Derivative: f'(x) = 1 - tanh^2(x)
- Pros:
    - Zero-centered output.
    - Differentiable.
- Cons:
    - Suffers from vanishing gradient problem (similar to sigmoid)."""
#Example code:
import numpy as np
import matplotlib.pyplot as plt

# Define activation functions
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

def relu(x):
    return np.maximum(0, x)

def tanh(x):
    return np.tanh(x)

# Plot activation functions
x = np.linspace(-5, 5, 100)
plt.plot(x, sigmoid(x), label='Sigmoid')
plt.plot(x, relu(x), label='ReLU')
plt.plot(x, tanh(x), label='Tanh')
plt.legend()
plt.show()
```

```python
#4. What is the difference between Loss function and Cost function in neural networks? Provide examples.
"""Loss Function and Cost Function are often used interchangeably in the context of neural networks, but there is a subtle difference between them.

Loss Function:- A loss function is a mathematical function that measures the difference between the predicted output and the actual output for a single data point or a single output.
- It is used to quantify the error or loss for a single example.
- The goal is to minimize the loss function for each data point.

Cost Function:- A cost function, also known as the objective function, is a mathematical function that measures the average loss over the entire training dataset.
- It is used to evaluate the performance of the model on the entire dataset.
- The goal is to minimize the cost function, which in turn minimizes the loss function for each data point.

Key differences:- Loss function is for a single data point, while cost function is for the entire dataset.
- Loss function is used to calculate the error for each data point, while cost function is used to calculate the average error over the entire dataset.

Examples:- Mean Squared Error (MSE) is a loss function that measures the squared difference between predicted and actual values for a single data point.
- Mean Squared Error Cost Function is the average MSE over the entire dataset.

Mathematically, if we have a dataset of n samples, the loss function for a single data point i can be represented as:

L_i = (y_i - y_pred_i)^2

The cost function (MSE) for the entire dataset would be:

J = (1/n) * Σ(L_i) = (1/n) * Σ(y_i - y_pred_i)^2

Other examples of loss functions and cost functions:- Cross-Entropy Loss (for classification problems)
- Binary Cross-Entropy Loss (for binary classification problems)
- Categorical Cross-Entropy Loss (for multi-class classification problems)"""

#Code Example (Python):
import numpy as np

# Define mean squared error loss function
def mse_loss(y_true, y_pred):
    return np.mean((y_true - y_pred) ** 2)

# Define mean squared error cost function
def mse_cost(y_true, y_pred):
    return np.mean(mse_loss(y_true, y_pred))

# Test the loss and cost functions
y_true = np.array([1, 2, 3])
y_pred = np.array([1.1, 2.1, 2.9])
print(mse_loss(y_true, y_pred))
print(mse_cost(y_true, y_pred))
```

```python
#5. What is the role of optimizers in neural networks? Compare Gradient Descent, Adam, and RMSprop.
""" ptimizers play a crucial role in neural networks by minimizing the loss function and updating the model's parameters to achieve optimal performance.

Role of Optimizers:
1. Minimize Loss: Optimizers aim to minimize the loss function, which measures the difference between the predicted and actual outputs.
2. Update Parameters: Optimizers update the model's parameters to reduce the loss and improve the model's performance.

Comparison of Gradient Descent, Adam, and RMSprop:
1. Gradient Descent (GD)- Update Rule: w = w - α * ∇L(w)
- Pros:
    - Simple to implement.
    - Converges to the optimal solution for convex problems.
- Cons:
    - Slow convergence for large datasets.
    - May get stuck in local minima.

2. Adam (Adaptive Moment Estimation)- Update Rule: w = w - α * m_t / sqrt(v_t + ε)
- Pros:
    - Adaptive learning rate for each parameter.
    - Combines benefits of RMSprop and momentum.
    - Fast convergence.
- Cons:
    - Requires more memory to store moment estimates.

3. RMSprop (Root Mean Square Propagation)- Update Rule: w = w - α * ∇L(w) / sqrt(v_t + ε)
- Pros:
    - Adaptive learning rate for each parameter.
    - Helps to stabilize the update process.
- Cons:
    - May not converge to the optimal solution for some problems.

Key differences:- Learning Rate Adaptation: Adam and RMSprop adapt the learning rate for each parameter, while Gradient Descent uses a fixed learning rate.
- Momentum: Adam incorporates momentum, which helps to escape local minima.
Choosing an Optimizer:- Adam is a popular choice for many deep learning applications.
- RMSprop can be used when memory is limited.
- Gradient Descent can be used for simple problems or when interpretability is important."""

#Example Code (Python):
import torch
import torch.nn as nn
import torch.optim as optim

# Define a simple neural network
class Net(nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = nn.Linear(5, 10)
        self.fc2 = nn.Linear(10, 5)

    def forward(self, x):
        x = torch.relu(self.fc1(x))
        x = self.fc2(x)
        return x

# Initialize the network and optimizer
net = Net()
criterion = nn.MSELoss()

# Gradient Descent
optimizer_gd = optim.SGD(net.parameters(), lr=0.01)

# Adam
optimizer_adam = optim.Adam(net.parameters(), lr=0.001)

# RMSprop
optimizer_rmsprop = optim.RMSprop(net.parameters(), lr=0.001)

# Train the network
for epoch in range(100):
    # Forward pass
    inputs = torch.randn(10, 5)
    labels = torch.randn(10, 5)
    outputs = net(inputs)
    loss = criterion(outputs, labels)

    # Backward pass
    optimizer_gd.zero_grad()
    optimizer_adam.zero_grad()
    optimizer_rmsprop.zero_grad()
    loss.backward()

    # Update parameters
    optimizer_gd.step()
    optimizer_adam.step()
    optimizer_rmsprop.step()
```

```python
#6. Write a Python program to implement a single-layer perceptron from scratch using NumPy to solve the logical AND gate.
#(Include your Python code and output in the code box below.)

import numpy as np

class Perceptron:
    def __init__(self, learning_rate=0.01, num_iterations=1000):
        self.learning_rate = learning_rate
        self.num_iterations = num_iterations
        self.weights = None
        self.bias = None

    def fit(self, X, y):
        num_samples, num_features = X.shape
        self.weights = np.zeros(num_features)
        self.bias = 0

        for _ in range(self.num_iterations):
            for idx, x_i in enumerate(X):
                linear_output = np.dot(x_i, self.weights) + self.bias
                y_predicted = np.where(linear_output >= 0, 1, 0)

                update = self.learning_rate * (y[idx] - y_predicted)
                self.weights += update * x_i
                self.bias += update

    def predict(self, X):
        linear_output = np.dot(X, self.weights) + self.bias
        y_predicted = np.where(linear_output >= 0, 1, 0)
        return y_predicted

# Define the dataset for the logical AND gate
X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
y = np.array([0, 0, 0, 1])

# Create a Perceptron object
perceptron = Perceptron()

# Train the model
perceptron.fit(X, y)

# Make predictions
predictions = perceptron.predict(X)

# Print the predictions
print("Predictions:")
for i, prediction in enumerate(predictions):
    print(f"Input: {X[i]} | Output: {prediction}")

print("\nFinal Weights:", perceptron.weights)
print("Final Bias:", perceptron.bias)
```

```python
#8. Use Keras to build and train a simple multilayer neural network on the MNIST digits dataset. Print the training accuracy.
#(Include your Python code and output in the code box below.)
import numpy as np
from tensorflow.keras.datasets import mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical

# Load the MNIST dataset
(x_train, y_train), (x_test, y_test) = mnist.load_data()

# Reshape the input data
x_train = x_train.reshape(-1, 784)
x_test = x_test.reshape(-1, 784)

# Normalize the input data
x_train = x_train.astype('float32') / 255
x_test = x_test.astype('float32') / 255

# Convert class labels to categorical labels
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Build the neural network model
model = Sequential()
model.add(Dense(512, activation='relu', input_shape=(784,)))
model.add(Dropout(0.2))
model.add(Dense(256, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(10, activation='softmax'))

# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
model.fit(x_train, y_train, batch_size=128, epochs=5, validation_data=(x_test, y_test))

# Evaluate the model
loss, accuracy = model.evaluate(x_test, y_test)
print(f'Test accuracy: {accuracy:.2f}')

# Get the training accuracy
train_loss, train_accuracy = model.evaluate(x_train, y_train)
print(f'Training accuracy: {train_accuracy:.2f}')
```

```python
#10. You are working on a project for a bank that wants to automatically detect fraudulent transactions. The dataset is large, imbalanced, and contains structured
#features like transaction amount, merchant ID, and customer location. The goal is to classify each transaction as fraudulent or legitimate.
#Explain your real-time data science workflow:
#● How would you design a deep learning model (perceptron or multilayer NN)?
#● Which activation function and loss function would you use, and why?
#● How would you train and evaluate the model, considering class imbalance?
#● Which optimizer would be suitable, and how would you prevent overfitting?
#(Include your Python code and output in the code box below.)
""" Fraudulent Transaction Detection using Deep Learning
Problem Statement:
The bank wants to detect fraudulent transactions in real-time using a deep learning model. The dataset is large, imbalanced, and contains structured features like transaction amount, merchant ID, and customer location.

Designing a Deep Learning Model:
We will design a multilayer neural network with the following architecture:

- Input layer with shape (n_features,)
- Hidden layer with 64 units, ReLU activation, and dropout rate 0.2
- Hidden layer with 32 units, ReLU activation, and dropout rate 0.2
- Output layer with 1 unit, sigmoid activation

Activation Function and Loss Function:
- We will use the ReLU activation function for the hidden layers because it is computationally efficient and helps the model learn complex relationships.
- We will use the sigmoid activation function for the output layer because it outputs a probability value between 0 and 1, which is suitable for binary classification problems.
- We will use the binary cross-entropy loss function because it is suitable for binary classification problems and measures the difference between predicted probabilities and actual labels.

Training and Evaluating the Model:
- We will use the Adam optimizer because it is a popular and efficient optimizer that adapts the learning rate for each parameter based on the magnitude of the gradient.
- We will use class weights to handle class imbalance. We will assign a higher weight to the minority class (fraudulent transactions) and a lower weight to the majority class (legitimate transactions).
- We will evaluate the model using metrics like precision, recall, F1-score, and area under the ROC curve (AUC-ROC).

Preventing Overfitting:
- We will use dropout regularization to prevent overfitting. Dropout randomly sets a fraction of neurons to zero during training, which helps the model learn more robust features.
- We will use early stopping to prevent overfitting. Early stopping stops the training process when the model's performance on the validation set starts to degrade."""

#Python Code:

import numpy as np
import pandas as pd # Import pandas
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from imblearn.over_sampling import SMOTE

# Load the dataset
# Assume that the dataset is stored in a Pandas dataframe called 'df'

# Create a synthetic dataset for demonstration purposes
np.random.seed(42)
num_samples = 1000
# Features: transaction_amount, merchant_id, customer_location (numeric for simplicity)
data = {
    'transaction_amount': np.random.rand(num_samples) * 1000,
    'merchant_id': np.random.randint(0, 100, num_samples),
    'customer_location': np.random.randint(0, 50, num_samples),
    'feature_4': np.random.rand(num_samples) * 10,
    'feature_5': np.random.rand(num_samples) * 5,
    'target': np.random.randint(0, 2, num_samples) # 0 for legitimate, 1 for fraudulent
}
df = pd.DataFrame(data)

# Introduce some imbalance and make some transactions fraudulent based on certain conditions
# For example, high transaction amounts might be more fraudulent
df.loc[df['transaction_amount'] > 900, 'target'] = 1
df.loc[df['merchant_id'] == 5, 'target'] = 1 # Specific merchant has more fraud

# Ensure some legitimate transactions even for high values to avoid perfect separation
df.loc[np.random.rand(num_samples) < 0.05, 'target'] = 0 # Randomly make some fraudulent legitimate

# Make the dataset imbalanced (e.g., 95% legitimate, 5% fraudulent)
num_fraudulent = int(num_samples * 0.05)
num_legitimate = num_samples - num_fraudulent

# Separate current fraudulent/legitimate transactions
df_fraud = df[df['target'] == 1]
df_legit = df[df['target'] == 0]

# Undersample legitimate if too many, or keep as is. For this example, let's just make sure fraud is a minority.
if len(df_fraud) < num_fraudulent:
    # If we don't have enough fraud from the conditions above, randomly add some
    additional_fraud_indices = np.random.choice(df_legit.index, num_fraudulent - len(df_fraud), replace=False)
    df.loc[additional_fraud_indices, 'target'] = 1

# Preprocess the data
X = df.drop('target', axis=1).values
y = df['target'].values

# Oversample the minority class using SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X, y)

# Split the data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42)

# Scale the data using StandardScaler
scaler = StandardScaler()
X_train = scaler.fit_transform(X_train)
X_val = scaler.transform(X_val)

# Build the model
model = Sequential()
model.add(Dense(64, activation='relu', input_shape=(X_train.shape[1],)))
model.add(Dropout(0.2))
model.add(Dense(32, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(1, activation='sigmoid'))

# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Train the model
history = model.fit(X_train, y_train, epochs=10, batch_size=128, validation_data=(X_val, y_val), verbose=2)

# Evaluate the model
y_pred = model.predict(X_val)
y_pred_class = (y_pred > 0.5).astype('int32')

print('Precision:', precision_score(y_val, y_pred_class))
print('Recall:', recall_score(y_val, y_pred_class))
print('F1-score:', f1_score(y_val, y_pred_class))
print('AUC-ROC:', roc_auc_score(y_val, y_pred))
```
